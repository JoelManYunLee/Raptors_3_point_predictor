```python
import pandas as pd 
import numpy as np 

from nba_api.stats.endpoints import BoxScoreAdvancedV3
```

```python
# Get stats from csv
box_score_df = pd.read_csv('boxscore_data.csv')
```

```python
# This function is not working yet

# Function to get advanced stats for a specific game
def get_advanced_stats(game_id):
    boxscore_advanced = BoxScoreAdvancedV3(game_id)
    player_stats = boxscore_advanced.player_stats.get_data_frame()
    return player_stats

# Function to iterate through games and create a DataFrame
def create_advanced_stats_dataframe(game_ids):
    player_stats_list = []

    for game_id in game_ids:
        player_stats = get_advanced_stats(game_id)
        player_stats_list.append(player_stats)

    all_player_stats = pd.concat(player_stats_list, ignore_index=True)

    return all_player_stats

# Extract the list of game IDs from the 'Game_ID' column of 'raptors_df'
game_ids_list = box_score_df['gameId'].tolist()

player_adv_stats_df = create_advanced_stats_dataframe(game_ids_list)
```

```python
box_score_df = box_score_df.dropna(subset=['three_point_field_goals_made'])

#box_score_df = box_score_df[box_score_df['season'] >= 2015]
box_score_df = box_score_df.sort_values('game_id', ascending=True)
box_score_df.head()
```

```python
# Adding the free_throw_percent column
box_score_df['free_throw_percent'] = box_score_df['free_throws_made'] / box_score_df['free_throws_attempted']
box_score_df = box_score_df.dropna(subset=['free_throw_percent'])
box_score_df.head()
```

```python
box_score_df_filtered = box_score_df[['game_id','game_date', 'season', 'season_type', 'athlete_display_name','team_name', 'team_id', 'minutes', 'three_point_field_goals_made', 'starter', 'free_throw_percent']]
box_score_df_filtered.head()
```

```python
# Group by 'team_name', 'game_id', and 'season'
#grouped_df = box_score_df_filtered.groupby(['game_id', 'team_name', 'team_id', 'season', 'game_date'])

#summary_data = grouped_df.agg(team_three_pm=('three_point_field_goals_made', 'sum')).reset_index()
#summary_data = summary_data.sort_values('game_date')

# Calculate cumulative mean and next game stats
#summary_data['three_pm_cumsum'] = summary_data.groupby(["team_name", "team_id", "season"])['team_three_pm'].cumsum()
#summary_data['game_count'] = summary_data.groupby(["team_name", "team_id", "season"]).cumcount() + 1
#summary_data['three_pm_avg_to_date'] = summary_data['three_pm_cumsum'] / summary_data['game_count']
#summary_data['team_three_pm_next_game'] = summary_data.groupby(["team_name", "team_id", "season"])['team_three_pm'].shift(-1)
#summary_data = summary_data.dropna(subset=['team_three_pm_next_game'])

# Now, sorted_df contains your desired DataFrame with grouping and sorting applied
#print(summary_data)

# Group by 'team_name', 'game_id', and 'season'
grouped_df = box_score_df_filtered.groupby(['game_id', 'team_name', 'team_id', 'season', 'game_date'])

summary_data = grouped_df.agg(
    team_free_throw_percent=('free_throw_percent', 'mean'),
    team_three_pm=('three_point_field_goals_made', 'sum')
).reset_index()
summary_data = summary_data.sort_values('game_date')

# Calculate cumulative mean and next game stats for 'free_throw_percent'
summary_data['free_throw_percent_cumsum'] = summary_data.groupby(["team_name", "team_id", "season"])['team_free_throw_percent'].cumsum()
summary_data['free_throw_percent_game_count'] = summary_data.groupby(["team_name", "team_id", "season"]).cumcount() + 1
summary_data['free_throw_percent_avg_to_date'] = summary_data['free_throw_percent_cumsum'] / summary_data['free_throw_percent_game_count']
summary_data['team_free_throw_percent_next_game'] = summary_data.groupby(["team_name", "team_id", "season"])['team_free_throw_percent'].shift(-1)

# Calculate cumulative mean and next game stats for 'three_point_field_goals_made'
summary_data['three_pm_cumsum'] = summary_data.groupby(["team_name", "team_id", "season"])['team_three_pm'].cumsum()
summary_data['three_pm_game_count'] = summary_data.groupby(["team_name", "team_id", "season"]).cumcount() + 1
summary_data['three_pm_avg_to_date'] = summary_data['three_pm_cumsum'] / summary_data['three_pm_game_count']
summary_data['team_three_pm_next_game'] = summary_data.groupby(["team_name", "team_id", "season"])['team_three_pm'].shift(-1)

# Drop rows with NaN values in 'team_free_throw_percent_next_game' and 'team_three_pm_next_game'
summary_data = summary_data.dropna(subset=['team_free_throw_percent_next_game', 'team_three_pm_next_game'])

# Now, summary_data contains your desired DataFrame with grouping, sorting, and cumulative mean applied
print(summary_data)
```
