```python
box_score_df = box_score_df.dropna(subset=['three_point_field_goals_made'])

# box_score_df = box_score_df[box_score_df['season'] >= 2010]
box_score_df = box_score_df.sort_values('game_id', ascending=True)
box_score_df.head()
```

```python
# Adding the free_throw_percent column
box_score_df['free_throw_percent'] = box_score_df['free_throws_made'] / box_score_df['free_throws_attempted']
box_score_df = box_score_df.dropna(subset=['free_throw_percent'])
box_score_df.head()
```

```python
# Adding the three point percent column
box_score_df['three_point_percent'] = box_score_df['three_point_field_goals_made'] / box_score_df['three_point_field_goals_attempted']
box_score_df = box_score_df.dropna(subset=['three_point_percent'])
```

```python
# Adding possessions column
box_score_df['possessions'] = box_score_df['field_goals_attempted'] + (0.44 * box_score_df['free_throws_attempted']) - box_score_df['offensive_rebounds'] + box_score_df['turnovers']
```

```python
# Adding orating column
box_score_df['orating'] = 100 * (box_score_df['points']/box_score_df['possessions'])
```

```python
# Filtering the dataframe for necessary columns
box_score_df_filtered = box_score_df[['game_id','game_date', 'season', 'season_type', 'athlete_display_name','team_name', 'team_id', 'opponent_team_id', 'minutes', 'points', 'field_goals_made', 'field_goals_attempted', 'three_point_field_goals_attempted','three_point_field_goals_made', 'three_point_percent', 'free_throw_percent', 'offensive_rebounds', 'defensive_rebounds', 'orating', 'turnovers']]
box_score_df_filtered.head()
```

```python
# Group by 'team_name', 'game_id', and 'season'
grouped_df = box_score_df_filtered.groupby(['game_id', 'team_name', 'team_id', 'season', 'game_date'])
grouped_df.head()
```

```python
# Aggregate data
summary_data = grouped_df.agg(
    team_fga=('field_goals_attempted', 'sum'),
    team_fgm=('field_goals_made', 'sum'),
    team_free_throw_percent=('free_throw_percent', 'mean'),
    team_three_pa=('three_point_field_goals_attempted', 'sum'),
    team_three_pm=('three_point_field_goals_made', 'sum'),
    team_three_percent=('three_point_percent', 'mean'),
    team_orating=('orating', 'mean'),
    team_oreb=('offensive_rebounds','sum'),
    team_dreb=('defensive_rebounds', 'sum'),
    team_to=('offensive_rebounds','sum'),
    team_points=('points','sum')
).reset_index()
summary_data = summary_data.sort_values('game_date')

summary_data.head()
```

```python
# List of columns to process
columns_to_process = ['team_fga', 'team_fgm', 'team_free_throw_percent', 'team_three_percent', 'team_three_pa', 'team_three_pm', 'team_orating', 'team_oreb', 'team_dreb', 'team_to', 'team_points']
    
def calculate_cumulative_stats(summary_data, column_name):
    # Calculate cumulative sum for the specified column
    summary_data[f'{column_name}_cumsum'] = summary_data.groupby(["team_name", "team_id", "season"])[column_name].cumsum()
    
    # Calculate game count
    summary_data[f'{column_name}_game_count'] = summary_data.groupby(["team_name", "team_id", "season"]).cumcount() + 1
    
    # Calculate average to date
    summary_data[f'{column_name}_avg_to_date'] = summary_data[f'{column_name}_cumsum'] / summary_data[f'{column_name}_game_count']
    
    # Calculate next game stats
    summary_data[f'{column_name}_next_game'] = summary_data.groupby(["team_name", "team_id", "season"])[column_name].shift(-1)
```

```python
# Process each column
for column in columns_to_process:
    calculate_cumulative_stats(summary_data, column)

# Add 'free_fries' column based on condition
summary_data['free_fries'] = summary_data['team_three_pm_next_game'] >= 12
    
# Drop rows with NaN values
summary_data.dropna(inplace=True)
```

```python
# Calculate rolling last five average
def calculate_last_five(summary_data, column_name, window=5):
    summary_data[f'{column_name}_last_5'] = summary_data.groupby(["team_name", "team_id", "season"])[column_name].rolling(window=window, min_periods=1).mean().reset_index(level=[0,1,2], drop=True)
    
    return summary_data

columns_last_5 = ['team_three_pa', 'team_three_pm', 'team_fga', 'team_fgm']

for column in columns_last_5:
    calculate_last_five(summary_data, column)
```
